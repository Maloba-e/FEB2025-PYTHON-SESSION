```python
#importing python libraries
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt

#loading data
data = pd.read_csv('Student_Performance_Data.csv')
data
```

```python
#displaying the first 5 rows of dataset
data.head()
```

```python
#getting the titles
data.columns
```

```python
#checking for missing value in the data set.
data.isnull()
```

```python
#computing basic statistic of the values in the dataset.
data.describe()
```

```python
#number of students
data['Student_ID'].nunique()
```

```python
#number of students who passed
data['Marks'].value_counts()
```

```python
#filtering the data

filtered_data = data[data['Marks']>80]
filtered_data
```

```python
#sorting the data
sorted_data = data.sort_values(by='Marks',ascending=False)
sorted_data
```

```python
#grouping the data
grouped_data = data.groupby('Student_ID')['Marks'].mean()
grouped_data
```

```python
#Aggregating the data
Avg_Marks = data['Marks'].mean()
Avg_Marks
```

```python
# Sampling the data
sampled_data = grouped_data.sample(n=10)
sampled_data
```

```python
#visualizing a sample of the data
sampled_data.plot(kind='line')
plt.title('Average Marks scored')
plt.ylabel('Marks')
plt.show()
```

```python
#visualizing a sample of the data
sampled_data.plot(kind='bar',color='green')
plt.title('Average Marks scored')
plt.ylabel('Marks')
plt.show()
```

```python
#visualizing a sample of the data
sampled_data.plot(kind='hist',color="black")
plt.title('Average Marks scored')
plt.ylabel('Marks')
plt.show()
```
